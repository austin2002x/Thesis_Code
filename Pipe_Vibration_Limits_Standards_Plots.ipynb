```python
plot_linspace = np.linspace(0,200,10000)

plot_WACHEL_DESIGN_displacement_PP_mils = (250/plot_linspace)**0.5
plot_WACHEL_MARGINAL_displacement_PP_mils = plot_WACHEL_DESIGN_displacement_PP_mils*2
plot_WACHEL_CORRECTION_displacement_PP_mils = plot_WACHEL_DESIGN_displacement_PP_mils*4
plot_WACHEL_DANGER_displacement_PP_mils = plot_WACHEL_DESIGN_displacement_PP_mils*10
plot_WACHEL_PERCEPTION_displacement_PP_mils = plot_WACHEL_DESIGN_displacement_PP_mils*0.25


plot_WACHEL_DESIGN_velocity_RMS_mm = plot_WACHEL_DESIGN_displacement_PP_mils*0.0254/(2*2**0.5)*2*np.pi*plot_linspace
plot_WACHEL_MARGINAL_velocity_RMS_mm = plot_WACHEL_MARGINAL_displacement_PP_mils*0.0254/(2*2**0.5)*2*np.pi*plot_linspace
plot_WACHEL_CORRECTION_velocity_RMS_mm = plot_WACHEL_CORRECTION_displacement_PP_mils*0.0254/(2*2**0.5)*2*np.pi*plot_linspace
plot_WACHEL_DANGER_velocity_RMS_mm = plot_WACHEL_DANGER_displacement_PP_mils*0.0254/(2*2**0.5)*2*np.pi*plot_linspace

plot_EI_CONCERN_velocity_RMS_mm = 10**((np.log10(plot_linspace)+0.48017)/(2.127612))
plot_EI_PROBLEM_velocity_RMS_mm = 10**((np.log10(plot_linspace)+1.871083)/(2.084547))

plot_EI_CONCERN_displacement_PP_mils = plot_EI_CONCERN_velocity_RMS_mm/0.0254*(2*2**0.5)/(2*np.pi*plot_linspace)
plot_EI_PROBLEM_displacement_PP_mils = plot_EI_PROBLEM_velocity_RMS_mm/0.0254*(2*2**0.5)/(2*np.pi*plot_linspace)
```

```python
fig, axes = plt.subplots(ncols=1, nrows=1, constrained_layout=True, figsize=(8, 6))

u=1.25/(2*np.pi*200)*1000

x_API618_mils=[0,10,200]
y_API618_mils=[20,20,u]

axes.set_title('Allowable Displacement Limits')
axes.set_xlim([0.5,200])
axes.loglog()
axes.set_xlabel('Frequency (Hz)')
axes.set_ylabel('Allowable Displacement (mils P-P)')

axes.plot(plot_linspace,plot_WACHEL_DESIGN_displacement_PP_mils, label='SWRI Design')
axes.plot(plot_linspace,plot_WACHEL_MARGINAL_displacement_PP_mils, label='SWRI Marginal')
axes.plot(plot_linspace,plot_WACHEL_CORRECTION_displacement_PP_mils, label='SWRI Correction')
axes.plot(plot_linspace,plot_WACHEL_DANGER_displacement_PP_mils, label='SWRI Danger')



axes.loglog()

add_inline_labels(axes, with_overall_progress=True, fontsize="small",debug=False)

axes.plot(x_API618_mils,y_API618_mils,'--',label="API 618")
handles, labels = axes.get_legend_handles_labels()
axes.legend(labels,markerscale=3)

#add_inline_labels(axes, with_overall_progress=True, fontsize="small",debug=False)
q=axes.get_lines()

labelLine(
    q[-1],
    5,
    label=r"{}".format(q[-1].get_label()),
    ha="center",
    va="center",
    align=True,
    backgroundcolor="none",
)

fig.show()
```
